```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
for sim, group in branches_df.groupby("Simulation"):
    plt.plot(group["Year"], group["Gross_Output"], alpha=0.3) 

mean_by_year = branches_df.groupby("Year")["Gross_Output"].mean()
plt.plot(mean_by_year.index, mean_by_year.values,
         color="blue", lw=2.5, label="Mean across simulations")

plt.xlabel("Year")
plt.ylabel("Gross Output")
plt.title("Monte Carlo Simulations")

plt.show()
```
