```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

realValAdd = pd.read_csv("spaceRealValueAdded.csv", )
row = realValAdd.loc[0]

row_list = [(k, row[k]) for k in row.keys()]
row_list_transpose = list(zip(*row_list))
years = [int(x) for x in row_list_transpose[0][1:]]
number = [int(x.replace(',', '')) for x in row_list_transpose[1][1:]]
# print(row)
```

```python
plt.xlabel("Year")
plt.ylabel("Real Value of Addition to Economy")
plt.plot(years, number)
```
